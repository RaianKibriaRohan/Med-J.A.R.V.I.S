# **What is EDA?**

EDA stands for **Exploratory Data Analysis**. It's the process of examining and understanding a dataset before you start modeling or drawing conclusions. EDA involves summarizing the main characteristics of the data using visual methods (like histograms, box plots, scatter plots) and statistical techniques (like mean, median, standard deviation, correlation, etc.).

**Why EDA is Important:**

*Understanding the Data:*

Before building any model, you need to know what you're dealing with. EDA helps you understand the structure, patterns, and relationships in your data.

*Identifying Data Quality Issues:*

It helps detect:

1. Missing values

2. Outliers

3. Duplicates

4. Inconsistent data types

*Guiding Feature Engineering:*

EDA can show which features might be useful, which ones are redundant, and how variables interact with each other. This insight is crucial for creating meaningful features.

*Detecting Trends and Patterns:*

EDA reveals trends, seasonality, or clusters that can influence modeling decisions.

*Avoiding Wrong Assumptions:*

Looking at the raw data can prevent mistakes like assuming normal distribution when the data is skewed.

*Improving Model Performance:*

Better understanding of your data often leads to better preprocessing, which leads to more accurate models.

**Basic EDA Pipeline**

*   Explore the dataset.
*   Understand its structure and contents.
*   Discover patterns, relationships and anomalies.
*   Prepare it for further modeling or decision-making.

###**Common Solutions to Class Imbalance:**
âœ… 1. Resampling Techniques


a. Oversampling (Minority Class)

Increases the number of samples in the minority class.

Tools: SMOTE, ADASYN, or simple random oversampling.

b. Undersampling (Majority Class)

Reduces the number of samples in the majority class.

Can risk losing important information.

c. Hybrid Methods

Combines both oversampling and undersampling for better balance.

Example: SMOTE-ENN, SMOTE-Tomek Links.

âœ… 2. Use of Class Weights

Assign higher penalty to misclassifying minority class. Many ML algorithms support this:

class_weight='balanced' in LogisticRegression, RandomForest, SVM, etc.

In deep learning: use weighted loss functions (e.g., weighted Cross Entropy).

âœ… 3. Synthetic Data Generation

Create artificial data points for minority class.

SMOTE (Synthetic Minority Over-sampling Technique) is the most popular method.

ADASYN is an advanced version of SMOTE that adapts based on the data distribution.

âœ… 4. Anomaly Detection Framing

If the minority class is extremely rare, consider treating it as an anomaly detection problem.

âœ… 5. Ensemble Methods

Bagging/Boosting techniques (e.g., Balanced Random Forest, XGBoost with scale_pos_weight) can perform well on imbalanced data.

These methods can learn patterns even in imbalanced settings by focusing on hard-to-classify instances.

âœ… 6. Data Augmentation (for Images, Text, etc.)

Apply transformations to minority class data to artificially expand it (for example, rotate/flip images).

âœ… 7. Evaluation Metrics Adjustment

Accuracy can be misleading in imbalanced data.

Use:

Precision, Recall, F1-Score

ROC-AUC

PR AUC (for very imbalanced datasets)

```python
# ==========================
# Google Colab Preprocessing Script
# ==========================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# ==========================
# 1. Load Dataset
# ==========================

df = pd.read_csv("medical_students_dataset.csv")

print("Original shape:", df.shape)


# ==========================
# 2. Drop rows with >40% missing
# ==========================
threshold = int(df.shape[1] * 0.6)   # at least 60% non-null required
df = df.dropna(thresh=threshold).reset_index(drop=True)
print("Shape after dropping rows with >40% missing:", df.shape)


# ==========================
# 3. Fill Student ID blanks with (prev + next)/2
# ==========================
if "Student ID" in df.columns:
    for i in range(len(df)):
        if pd.isnull(df.loc[i, "Student ID"]):
            prev_val = df.loc[i-1, "Student ID"] if i > 0 else np.nan
            next_val = df.loc[i+1, "Student ID"] if i < len(df)-1 else np.nan
            if not pd.isnull(prev_val) and not pd.isnull(next_val):
                df.loc[i, "Student ID"] = (prev_val + next_val) / 2
    # If still NaN (first or last row missing), fill with mean
    df["Student ID"].fillna(df["Student ID"].mean(), inplace=True)


# ==========================
# 4. Numeric columns fill with mean
# ==========================
num_cols = ["Age", "Height", "Weight", "BMI",
            "Temperature", "Heart Rate", "Blood Pressure", "Cholesterol"]

for col in num_cols:
    if col in df.columns:
        df[col].fillna(df[col].mean(), inplace=True)


# ==========================
# 5. Categorical: Diabetes, Smoking â†’ fill with majority class
# ==========================
for col in ["Diabetes", "Smoking"]:
    if col in df.columns:
        majority = df[col].mode()[0]
        df[col].fillna(majority, inplace=True)
        df[col] = df[col].map({"Yes": 1, "No": 0})


# ==========================
# 6. Gender: fill with majority, encode Male=1, Female=0
# ==========================
if "Gender" in df.columns:
    majority_gender = df["Gender"].mode()[0]
    df["Gender"].fillna(majority_gender, inplace=True)
    df["Gender"] = df["Gender"].map({"Male": 1, "Female": 0})


# ==========================
# 7. Blood Type â†’ One-hot encoding (A, B, AB, O)
# ==========================
if "Blood Type" in df.columns:
    majority_bt = df["Blood Type"].mode()[0]
    df["Blood Type"].fillna(majority_bt, inplace=True)

    # One-hot encode manually
    df["BloodType_A"] = (df["Blood Type"] == "A").astype(int)
    df["BloodType_B"] = (df["Blood Type"] == "B").astype(int)
    df["BloodType_AB"] = (df["Blood Type"] == "AB").astype(int)
    df["BloodType_O"] = (df["Blood Type"] == "O").astype(int)

    # Drop original column
    df.drop(columns=["Blood Type"], inplace=True)


# ==========================
# 8. StandardScaler for selected numeric cols
# ==========================
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("\nFinal shape after preprocessing:", df.shape)


# ==========================
# 9. Save cleaned dataset
# ==========================
df.to_csv("medical_students_dataset_cleaned.csv", index=False)
print("âœ… Cleaned dataset saved as 'medical_students_dataset_cleaned.csv'")
```

# **RANDOM SPLITTING OF CLEANSED DATA**

```python
# ==========================
# General Dataset Splitting (Train/Test)
# ==========================
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the cleaned dataset
df_clean = pd.read_csv("medical_students_dataset_cleaned.csv")

print("Clean dataset shape:", df_clean.shape)

# Split entire dataset into 70% train and 30% test
train_set, test_set = train_test_split(
    df_clean, test_size=0.3, random_state=42
)

# Save train and test sets
train_set.to_csv("train_dataset.csv", index=False)
test_set.to_csv("test_dataset.csv", index=False)

print("âœ… Train set shape:", train_set.shape)
print("âœ… Test set shape:", test_set.shape)
print("Files saved as 'train_dataset.csv' and 'test_dataset.csv'")
```


# **Model Training:**


```python
# ==========================
# Model Training & Testing with train/test CSVs
# Focus: Blood Pressure Relationship (BP_Class prediction)
# ==========================
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score

# -------------------------
# 1. Load train and test sets
# -------------------------
train_df = pd.read_csv("train_dataset.csv")
test_df = pd.read_csv("test_dataset.csv")

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# -------------------------
# 2. Recreate BP_Class (quantile split, 3 categories)
# -------------------------
q1 = train_df["Blood Pressure"].quantile(0.33)
q2 = train_df["Blood Pressure"].quantile(0.66)

def categorize_bp(bp):
    if bp < q1:
        return 0   # Low
    elif bp < q2:
        return 1   # Medium
    else:
        return 2   # High

train_df["BP_Class"] = train_df["Blood Pressure"].apply(categorize_bp)
test_df["BP_Class"] = test_df["Blood Pressure"].apply(categorize_bp)

print("\nTrain class distribution:\n", train_df["BP_Class"].value_counts())
print("\nTest class distribution:\n", test_df["BP_Class"].value_counts())

# -------------------------
# 3. Features & Target
# -------------------------
drop_cols = ["Blood Pressure", "BP_Class", "Student ID"]  # remove ID + BP itself
X_train = train_df.drop(columns=drop_cols)
y_train = train_df["BP_Class"]

X_test = test_df.drop(columns=drop_cols)
y_test = test_df["BP_Class"]

# âœ… Already standardized in preprocessing
X_train_scaled = X_train.values
X_test_scaled  = X_test.values

# -------------------------
# 4. Initialize Models (only DT, NN, KNN)
# -------------------------
models = {
    "Decision Tree": DecisionTreeClassifier(
        max_depth=12, min_samples_split=10,
        class_weight="balanced", random_state=42
    ),
    "Neural Network": MLPClassifier(
        hidden_layer_sizes=(128,64,32), activation="relu",
        solver="adam", max_iter=800, random_state=42
    ),
    "KNN": KNeighborsClassifier(n_neighbors=7)
}

# -------------------------
# 5. Train & Evaluate (Supervised)
# -------------------------
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    preds = model.predict(X_test_scaled)

    acc = accuracy_score(y_test, preds)
    print(f"\n=== {name} ===")
    print("Accuracy:", round(acc, 4))
    print("Classification Report:\n", classification_report(y_test, preds))

    cm = confusion_matrix(y_test, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Low","Medium","High"])
    disp.plot(cmap="Blues", values_format="d")
    disp.ax_.set_title(f"{name} - Confusion Matrix")

# -------------------------
# 6. Unsupervised KMeans
# -------------------------
print("\n=== KMeans Clustering (Unsupervised) ===")
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_test)

# Compare clusters to true BP_Class
from scipy.stats import mode
labels = np.zeros_like(clusters)
for i in range(3):
    mask = (clusters == i)
    if np.any(mask):
        labels[mask] = mode(y_test[mask], keepdims=True).mode[0]

acc_kmeans = accuracy_score(y_test, labels)
ari = adjusted_rand_score(y_test, clusters)
sil = silhouette_score(X_test, clusters)

print(f"KMeans Accuracy (matched to BP_Class): {acc_kmeans:.4f}")
print(f"Adjusted Rand Index: {ari:.4f}")
print(f"Silhouette Score: {sil:.4f}")

cm = confusion_matrix(y_test, labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Low","Medium","High"])
disp.plot(cmap="Purples", values_format="d")
disp.ax_.set_title("KMeans - Confusion Matrix vs BP_Class")
```

# **Model Evaluation:**

```python
# ==========================
# Model Evaluation (with train/test CSVs)
# ==========================
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
    classification_report, confusion_matrix, roc_auc_score, roc_curve
)
from sklearn.preprocessing import label_binarize

results = []
all_preds = {}

# Binarize labels for multi-class ROC
y_test_bin = label_binarize(y_test, classes=[0,1,2])

# Train & Evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None

    acc = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds, average="weighted")
    prec = precision_score(y_test, preds, average="weighted", zero_division=0)
    rec = recall_score(y_test, preds, average="weighted", zero_division=0)
    auc = roc_auc_score(y_test_bin, probs, multi_class="ovr") if probs is not None else None

    results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1 Score": f1,
        "AUC": auc
    })

    all_preds[name] = (preds, probs)

    # Confusion Matrix
    cm = confusion_matrix(y_test, preds)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=["Low","Medium","High"],
                yticklabels=["Low","Medium","High"])
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    print(f"\n=== {name} ===")
    print("Accuracy:", round(acc,4))
    print("F1 Score:", round(f1,4))
    print("Classification Report:\n", classification_report(y_test, preds))

# Results DataFrame
results_df = pd.DataFrame(results)
print("\n=== Performance Comparison ===")
print(results_df)

# Bar Chart - Accuracy and F1
results_df_plot = results_df.melt(id_vars="Model", value_vars=["Accuracy", "F1 Score"],
                                  var_name="Metric", value_name="Score")

plt.figure(figsize=(6,4))
sns.barplot(x="Model", y="Score", hue="Metric", data=results_df_plot)
plt.title("Model Comparison - Accuracy & F1 Score")
plt.ylim(0,1)
plt.show()

# ROC Curve for all models (One-vs-Rest)
plt.figure(figsize=(6,4))
for name, (preds, probs) in all_preds.items():
    if probs is not None:
        for i, label in enumerate(["Low","Medium","High"]):
            fpr, tpr, _ = roc_curve(y_test_bin[:,i], probs[:,i])
            auc = roc_auc_score(y_test_bin[:,i], probs[:,i])
            plt.plot(fpr, tpr, label=f"{name} - {label} (AUC={auc:.2f})")

plt.plot([0,1],[0,1],"--",color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves (Multi-Class)")
plt.legend()
plt.show()

```

# **Conclusion**
In this project, we attempted to predict blood pressure categories (Low, Medium, High) using medical and lifestyle attributes such as BMI, cholesterol, smoking, age, height, and gender. Three supervised models (KNN, Decision Tree, Neural Network) and one unsupervised approach (KMeans clustering) were applied.

Why we used these models:

**Decision Tree â†’** interpretable, shows which features (e.g., BMI, cholesterol) most influence blood pressure.

**Neural Network â†’** mandatory in the guideline, captures non-linear relationships between features and BP.

**KNN â†’** good for similarity-based classification, often performs well in health-related datasets.

**KMeans â†’** required unsupervised method, used to check if natural clustering of BP categories exists.

Reasons behind low accuracy **(~33â€“38%)**:

**Weak correlations:** Features like BMI, cholesterol, and smoking influence blood pressure but are not strong predictors on their own. Blood pressure also depends on other factors (diet, exercise, genetics, stress) not present in the dataset.

**Quantile-based target:** BP was divided into 3 equal groups (Low/Medium/High) using quantiles, which may not reflect true medical categories and introduced noise.

**Feature overlap:** Distributions of BMI, cholesterol, and smoking overlap across all BP classes, making separation difficult for models.

**Clustering limitation:** KMeans assumes clear spherical clusters, but BP values do not cluster naturally â€” explaining the poor silhouette score and low adjusted rand index.

**Best Model:**

KNN gave the highest accuracy **(~37.9%)** with relatively balanced predictions.

Neural Network was second best **(~36.3%)** and performed consistently across classes.

Decision Tree was less accurate but provided useful interpretability.
